```python
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['pdf.use14corefonts'] = True

window_pre = 4
window_post = 10
name = 'NTA'
axis = None
cmap = plt.cm.viridis
def plot_switches(tape, cfg, ax, window_pre=4, window_post=10, name=None, axis=None, cmap = plt.cm.viridis):
    _tape = jax.tree_map(lambda x: np.array(x)[:], tape) if tape.t.ndim == 2 else tape
    ti_switches = np.where(np.abs(np.gradient(_tape.c_gt1[0]) / cfg.dt_tape) > 0.1)[0]
    ti_switches= ti_switches[::2]
    if name == 'flexible':
        ti_switches = ti_switches[0:4]
    else:
        ti_switches = ti_switches[0:4] 
    switch_centered_losses_means = []
    switch_centered_losses_sem = []
    for i, t_switch in enumerate(ti_switches.tolist()):
        if t_switch > window_pre and t_switch < ((_tape.loss.shape[1])-window_post):
            switch_centered_means = np.mean(_tape.loss[:, t_switch-window_pre:t_switch+window_post], axis=0)
            swtich_centered_sem = np.std(_tape.loss[:, t_switch-window_pre:t_switch+window_post], axis=0)
            switch_centered_losses_means.append(switch_centered_means)
            switch_centered_losses_sem.append(swtich_centered_sem)

    switch_losses_means  = np.array(switch_centered_losses_means)
    switch_losses_sem = np.array(switch_centered_losses_sem)

    total_switches = len(switch_losses_means)
    for i, (loss_m, loss_sem) in enumerate(zip(switch_losses_means, switch_losses_sem)):
        ax.plot(range(1,6),  loss_m[window_pre+2:window_pre+7], color=cmap(i/len(ti_switches)), label=f'switch {int(i*(i+1)+2)}' , alpha=0.8, marker='o', linestyle='-')
        if name == 'forgetful':
            loss_sem *= np.sqrt(20) 
        ax.errorbar(range(1,6), loss_m[window_pre+2:window_pre+7], yerr=loss_sem[window_pre+2:window_pre+7], color=cmap(i/len(ti_switches)), alpha=0.8)

    ax.set_xlabel(r'timesteps $t$ after switch')
    ax.set_ylabel(r'loss $\mathcal{L}$')
    ax.set_title(f'{name if name else ""}', fontsize=10)
    if name == 'forgetful':
        ax.legend(['blocks trained {}'.format(b) for b in [1,2,3,4]], fontsize=8, bbox_to_anchor=(1., 1.)) #, bbox_to_anchor=(.3, 0.6))

    ax.set_xlim([.5, 6])
fig, axes = plt.subplots(1, 2, figsize=(4.5, 1.8), sharey=True)
ax = axes[0]
plot_switches(tape, NTA_cfg, ax, window_pre, window_post, name='flexible', cmap=plt.cm.viridis)
ax.set_xticks(range(1,6))
ax = axes[1]
ax.set_xticks(range(1,6))
ax.invert_yaxis()

plot_switches(tape_base, forget_NTA_cfg, ax, window_pre, window_post, name='forgetful', cmap=plt.cm.viridis)
fig
```
