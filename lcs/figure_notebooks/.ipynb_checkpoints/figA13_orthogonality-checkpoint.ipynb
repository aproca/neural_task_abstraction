```python
logger = logging.getLogger('lcs')
logger.setLevel(logging.INFO)

g = 100  # hardness of switch

output_size = 2
Y1_tgt, Y2_tgt = jnp.eye(2)[:2]
Y_tgts = jnp.stack([Y1_tgt, Y2_tgt], axis=0)

overrides = dict(num_seeds=1,
                T_tape=-1,
                dt=0.01,
                teacher_mode="orthogonal",
                c_gt_curriculum='A_B__',  
                Y_tgt=None,
                regularization_strength=0.1,
                regularization_type=[('nonnegative', 0), ('gating_manifold_L2', 10.), ("L2_W", 0.)],
                initialization_scale=.01, 
                )
cfg_dict = dict()
```

```python
xx = np.linspace(0.0, 1.0, 5) #20

cfgs = [Config(**dict(cfg_lcs_dict_ | overrides, teacher_xx=xx)) for xx in xx]
tapes = [run_config(cfg) for cfg in tqdm(cfgs)]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4, 2), layout='constrained')
# show axes spines

ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
lb = len(tapes[0]['cos_sim1']) // 8

def get_adapt(cossim, lb=1):
    cossim = cossim[-lb:]
    diag = np.eye(2).astype(bool)

    # low_pass_filter cossim
    filter_size = len(cossim) // 10
    convolve = partial(jax.numpy.convolve, v=np.ones(filter_size) / filter_size, mode='same')
    cossim = jax.numpy.array(cossim)
    convolve = jax.vmap(convolve, in_axes=(1))
    convolve = jax.vmap(convolve, in_axes=(2))
    cossim = convolve(cossim)

    return np.abs(cossim[diag]- cossim[~diag]).max()

adapts_c = [np.mean([get_adapt(tape['cos_sim1'][i], lb) for i in range(cfg.num_seeds)], axis=0) for tape in tapes]

@partial(np.vectorize, signature='(t),(t),(),()->(n),(n)')
def get_first_crossing_times(loss, c_gt1, cfg, thsd=0.1):
    ti_switches = np.where(np.abs(np.gradient(c_gt1) / cfg.dt_tape) > 0.2)[0] + 1

    losses_blocks = np.split(loss, ti_switches)
    losses_blocks = [loss_block for loss_block in losses_blocks if len(loss_block) > 2]
    blocks = np.arange(1, len(losses_blocks)+1)

    loss_grads = []
    for loss_block in losses_blocks:
        loss_grad = np.abs(np.gradient(loss_block) / cfg.dt_tape).max()
        loss_grads.append(loss_grad)

    loss_grads = np.array(loss_grads)
    return blocks, loss_grads

get_first_crossing_times_ = partial(get_first_crossing_times, thsd=.2)

loss_grads = []
for tape in tapes:
    blocks, loss_grad = get_first_crossing_times_(tape.loss, tape.c_gt1, cfg)
    
    loss_grads.append(np.mean(loss_grad, axis=1).squeeze())

ax2 = ax.twinx()

ax2.plot(xx, adapts_c, color='gray')

ax.plot(xx, loss_grads, c='k')
ax.set_xlabel(r"teacher correlation $\boldsymbol{W}^m \cdot \boldsymbol{W}^{m'}$")
str_ = "student specialization\n" + r"$\boldsymbol{W}^{p\star} \cdot \boldsymbol{W}^p - \boldsymbol{W}^{p\star} \cdot \boldsymbol{W}^{p'}$"
ax2.set_ylabel(str_, color='gray')
ax.set_ylim(0, None)
ax.set_ylabel("adaptation speed (loss)")

fig
```
