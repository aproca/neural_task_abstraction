```python
import numpy as np
import equinox as eqx
import jax.numpy as jnp
import os
os.environ['ENABLE_PJRT_COMPATIBILITY'] = '1'
import jax
jax.config.update('jax_platform_name', 'cpu')
jax.config.update('jax_enable_x64', True)
import os
os.environ['JAX_PLATFORMS'] = 'cpu'
from functools import partial
import jax.numpy as jnp
import matplotlib.pyplot as plt
from lcs import PAGEHEIGHT, TEXTWIDTH
from matplotlib import colors
from lcs.plotting_utils import setup_axes_labels
```

```python
@partial(jnp.vectorize, signature='(i),(i),(p)->(i,j)')
def NTK(w1, w2, c):

    w = jnp.stack([w1, w2], axis=0)
    I = jnp.eye(w1.shape[-1])
    NTK_ = jnp.einsum('pi,pj->pij', w, w) + I*(c**2)[:, None, None]

    NTK = NTK_.sum(axis=0)
    return NTK

eps = jnp.array([-1, 1])

w11 = jnp.linspace(0, 1, 100)
w12 = jnp.linspace(0, 1, 100)

W11, W12 = jnp.meshgrid(w11, w12)

W21 = 1 - W11
W22 = 1 - W12

W1 = jnp.stack([W11, W12], axis=-1)
W2 = jnp.stack([W21, W22], axis=-1)

C = jnp.full(W11.shape + (2,), .5)

# instead, choose c to be .5, .5 in the center and 0, 1 at the edges
C0 = W1 @ jnp.array([1, 0])
C1 = W1 @ jnp.array([0, 1])
C = jnp.stack([C0, C1], axis=-1)

NTK_BOTH = NTK(W1, W2, C)
NTK_W = NTK(W1, W2, 0.*C)
NTK_C = NTK_BOTH - NTK_W

DY_BOTH = jnp.einsum('...ij,j->...i', NTK_BOTH, eps)
DOT_PROD_BOTH = jnp.einsum('i,...i->...', eps, DY_BOTH)

DY_C = jnp.einsum('...ij,j->...i', NTK_C, eps)
DOT_PROD_C = jnp.einsum('i,...i->...', eps, DY_C)

DY_W = jnp.einsum('...ij,j->...i', NTK_W, eps)
DOT_PROD_W = jnp.einsum('i,...i->...', eps, DY_W)

DOT_PROD_W = DOT_PROD_W.at[W11 + W12 > 1.].set(jnp.nan)
DOT_PROD_BOTH = DOT_PROD_BOTH.at[W11 + W12 > 1.].set(jnp.nan)
DOT_PROD_C = DOT_PROD_C.at[W11 + W12 > 1.].set(jnp.nan)

print(DOT_PROD_BOTH.shape)

s=2
fig, (axd) = plt.subplot_mosaic("ABC", figsize=(s*TEXTWIDTH, s*PAGEHEIGHT*.2), layout='constrained')    
ax1, ax2, ax3 = axd.values()

# make a logistic norm
def _forward(x):
    return np.tanh(x)


def _inverse(x):
    return np.arctanh(x)

# norm = colors.FuncNorm((_forward, _inverse), vmin=0, vmax=np.max([DOT_PROD_BOTH, DOT_PROD_C, DOT_PROD_W]))
norm = colors.PowerNorm(gamma=.5, vmin=0, vmax=np.nanmax([DOT_PROD_BOTH, DOT_PROD_C, DOT_PROD_W]))

sm = ax1.pcolormesh(w11, w12, DOT_PROD_BOTH, cmap='coolwarm', shading='auto', alpha=.5, norm=norm, rasterized=True)
ax1.set_box_aspect(1)
ax1.set_xlabel('$w_1^p$')
ax1.set_ylabel('$w_2^p$')
ax1.set_title('both terms')

sm = ax2.pcolormesh(w11, w12, DOT_PROD_W, cmap='coolwarm', shading='auto', alpha=.5, norm=norm, rasterized=True)
ax2.set_box_aspect(1)
ax2.set_xlabel('$w_1^p$')
ax2.set_ylabel('$w_2^p$')
ax2.set_title(r'$\sum_p \boldsymbol{w}^p \boldsymbol{w}^{pT}$ contribution')

ax3.set_title(r'$\sum_p c^p c^p$ contribution')
sm = ax3.pcolormesh(w11, w12, DOT_PROD_C, cmap='coolwarm', shading='auto', alpha=.5, norm=norm, rasterized=True)
ax3.set_box_aspect(1)
ax3.set_xlabel('$w_1^p$')
ax3.set_ylabel('$w_2^p$')

# make arrows

wbar = jnp.linspace(0.1, .4, 3)

for ax in [ax1, ax2, ax3]:
    for wbar_ in wbar:
        w1 = jnp.array([.5 - wbar_, .5 + wbar_])
        w2 = jnp.array([.5 + wbar_, .5 - wbar_])

        ax.arrow(0, 0, w1[0], w1[1], head_width=0.05, head_length=0.1/1.5, fc='C0', ec='C0', length_includes_head=True)
        ax.arrow(0, 0, w2[0], w2[1], head_width=0.05, head_length=0.1/1.5, fc='C1', ec='C1', length_includes_head=True)
        ax.axline((0, 1), (1, 0), color='k', linestyle='--')

fig.colorbar(sm, ax=ax3, label=r'$(\frac{d}{dt} \boldsymbol{y})^T \boldsymbol{\varepsilon}$', shrink=.5)
fig
```
